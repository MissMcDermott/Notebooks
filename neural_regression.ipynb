In this notebook we examine using neural networks for regression

```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import r2_score
```

We'll start with the student performance data we used in the multiple independent linear regression notebook

```python
df = pd.read_csv('data/Student_Performance.csv')
df.head()
```

```python
df.info()
```

As a referesher, this is how our inependent variables related to the performance index

```python
features = ['Hours Studied', 'Previous Scores', 'Extracurricular Activities', 'Sleep Hours', 'Sample Question Papers Practiced']

plt.subplots(figsize=(20,10))

for i, col in enumerate(features):
  plt.subplot(2,3,i+1)
  plt.title(features[i])
  plt.plot(df[col], df['Performance Index'])
plt.show()
```

We can also try using our heatmap to compare. First we will need to map the extracurricular activities to a boolean rather than strings

```python
df['Extracurricular Activities'] = df['Extracurricular Activities'].map({'Yes': True, 'No': False})
sb.heatmap(df.corr())
```

```python
X = df.drop("Performance Index", axis=1)
y = df['Performance Index']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=101)
```

```python
scaler = StandardScaler()
scaler.fit(X_train)

X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
rgr = MLPRegressor(solver='lbfgs', alpha=1e-5, hidden_layer_sizes=(5, 5), random_state=1)

rgr.fit(X_train, y_train)
```

```python
predictions = rgr.predict(X_test)

print(r2_score(y_test, predictions))
```
