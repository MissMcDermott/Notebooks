In this file we will implement and visualise KNN models. We will just be generating our own data so that we can actually visualise it

We have to do our imports as per usual

```python
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import datasets, neighbors
from sklearn.model_selection import train_test_split
from mlxtend.plotting import plot_decision_regions
```

Our generation is below

```python
X, y = datasets.make_blobs(n_samples=500, n_features=2, centers=4, cluster_std=1.5, random_state=4)
```

```python
knn = neighbors.KNeighborsClassifier(n_neighbors= 5)
knn.fit(X, y)
```

```python
plot_decision_regions(X, y, clf=knn, legend=2)
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
